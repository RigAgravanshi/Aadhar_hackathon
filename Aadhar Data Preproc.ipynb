```python
def process(df):   #pass dataframe
    df["state"] = df["state"].str.upper()
    df['state'] = df['state'].replace(r'\s+', ' ', regex=True)
    return df
```

```python
df_combined = pd.concat([df1, df2, df3], axis=0, ignore_index=True, verify_integrity=True)
```

```python
df_combined
```

```python
df_combined["state"].nunique()
```

```python
df_process = process(df_combined)
```

```python
df_process["state"].value_counts()
```

```python
df_process["state"].nunique()
```

```python
state_mapping = {
    # Misspellings
    "WESTBENGAL" : "WEST BENGAL",
    "WEST BANGAL" : "WEST BENGAL",
    "JAMMU & KASHMIR": "JAMMU AND KASHMIR",
    "ANDAMAN & NICOBAR ISLANDS" : "ANDAMAN AND NICOBAR ISLANDS",

    # Name updates
    "ORISSA": "ODISHA",
    "PONDICHERRY": "PUDUCHERRY",
    
    "THE DADRA AND NAGAR HAVELI AND DAMAN AND DIU" : "DADRA AND NAGAR HAVELI AND DAMAN AND DIU",
    "DAMAN AND DIU" : "DADRA AND NAGAR HAVELI AND DAMAN AND DIU",
    "DAMAN & DIU" : "DADRA AND NAGAR HAVELI AND DAMAN AND DIU",
    "DADRA AND NAGAR HAVELI" : "DADRA AND NAGAR HAVELI AND DAMAN AND DIU",
    "DADRA & NAGAR HAVELI" : "DADRA AND NAGAR HAVELI AND DAMAN AND DIU",

    # The unnamed states
    "100000" : "UNKNOWN"
}
```

```python
df_process["state"] = df_process["state"].replace(state_mapping)
```

```python
df_process["state"].nunique()   # Should be equal to 28(states) + 8(UT) +1(unknown)
```
