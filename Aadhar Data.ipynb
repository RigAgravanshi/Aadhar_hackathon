```python
def process(df):   #pass dataframe
    df["state"] = df["state"].str.upper()
    df['state'] = df['state'].replace(r'\s+', ' ', regex=True)
    return df
```

```python
df_combined = pd.concat([df1, df2, df3], axis=0, ignore_index=True, verify_integrity=True)
```

```python
df_combined["state"].nunique()
```

```python
df_process = process(df_combined)
```
